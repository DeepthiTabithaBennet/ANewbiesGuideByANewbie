```python
import transportation_problem as tp
import gradio as gr
```

```python
def result1(s, d, c):

    s = eval(s)
    d = eval(d)
    c = eval(c)

    p  =  tp.TransportationProblem(s, d, c)
    r  =  p.solve(tp.VogelIniter)
    return(r)
```

```python
iface = gr.Interface(
    fn=result1,
    inputs=[
        gr.Textbox(label="Supply Values, in Array of Tuples Format"),
        gr.Textbox(label="Demand Values, in Array of Tuples Format"),
        gr.Textbox(label="Cost Values, in Matrix Format")
    ],
    allow_flagging="never",
    outputs="text",
    title="Vogel's Approximation Method (Without MODI) Solver"
)

iface.launch()
```

```python
def result2(s, d, c):

    s = eval(s)
    d = eval(d)
    c = eval(c)

    p  =  tp.TransportationProblem(s, d, c)
    r  =  p.solve(tp.VogelIniter, tp.PotentialChecker, tp.ClosedLoopAdjustmentOptimizer)
    return(r)
```

```python
iface = gr.Interface(
    fn=result2,
    inputs=[
        gr.Textbox(label="Supply Values, in Array of Tuples Format"),
        gr.Textbox(label="Demand Values, in Array of Tuples Format"),
        gr.Textbox(label="Cost Values, in Matrix Format")
    ],
    allow_flagging="never",
    outputs="text",
    title="Vogel's Approximation Method (With MODI) Solver"
)

iface.launch()
```
