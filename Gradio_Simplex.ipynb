```python
from scipy.optimize import linprog
import numpy as np
import gradio as gr
```

```python
def result(c_vec, a_matrix, b_vec):

    obj = eval(c_vec)
    lhs = eval(a_matrix)
    rhs = eval(b_vec)

    simplex = linprog(c = obj,
                  A_ub = lhs,
                  b_ub = rhs,
                  method = 'simplex')
    return(simplex)
```

```python
iface = gr.Interface(
    fn=result,
    inputs=[
        gr.Textbox(label="Objective Function Coefficients, in Array Format (OBJ)"),
        gr.Textbox(label="Constraint Coefficients, in Matrix Format (LHS)"),
        gr.Textbox(label="Constraints Values, in Array Format (RHS)")
    ],
    allow_flagging="never",
    outputs="text",
    title="Simplex Method Solver"
)

iface.launch()
```
